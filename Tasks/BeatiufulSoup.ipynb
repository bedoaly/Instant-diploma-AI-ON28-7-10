```python
import requests
from bs4 import BeautifulSoup
#some import lib
url_of_website = 'https://books.toscrape.com/catalogue/page-{}.html'
#this will keep the url of website in variable url_of_website
page = 1 #count for number of page
while True:
    url = url_of_website.format(page)
    response = requests.get(url)
    if response.status_code != 200:
        break
    #sends an HTTP GET request to the URL 
    content = response.content
    #her the content
    soup = BeautifulSoup(content, 'html.parser')
    books = soup.find_all('article', class_='product_pod')
    for book in books:
        title = book.h3.a['title']
        #extracts the title of book
        price = book.select('div p.price_color')[0].text
        rating = 'No rating' if 'star-rating' not in str(book) else book.select('p.star-rating')[0]['class'][1]
        #if there is no rating for a book     
        if rating=="Five":
            print(f"The title of book = {title}")
            print(f"The price of book =  {price}")
            print(f"The rating for book out of 5 = {rating}")
            print("----------")
        # print informations
    page = page + 1
print("print all bookes with five stars")
```
